```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def booth(x1, x2):
	return (x1 + 2 * x2 - 7) ** 2 + (2 * x1 + x2 - 5) ** 2
```

```python
X_1d = np.linspace(-10, 10, 100)
Y_1d= np.linspace(-10, 10, 100)

X, Y = np.meshgrid(X_1d, Y_1d)
Z = booth(X, Y)

plt.figure(figsize=(8, 8))
contourf = plt.contourf(X, Y, Z, levels=20)
plt.colorbar(contourf, label='Z value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Booth Function')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

```python
X_flat = X.flatten()
Y_flat = Y.flatten()
Z_flat = booth(X_flat, Y_flat)

num_examples = len(X_flat)
num_inputs = 2

input_filename = "input.txt"
output_filename = "output.txt"

with open(input_filename, "w") as f_input:
	f_input.write(f"{num_examples} {num_inputs}\n")
	for i in range(num_examples):
		f_input.write(f"{X_flat[i]:.6f} {Y_flat[i]:.6f}\n")

with open(output_filename, "w") as f_output:
	f_output.write(f"{num_examples}\n")
	for i in range(num_examples):
		f_output.write(f"{Z_flat[i]:.6f}\n")
```
